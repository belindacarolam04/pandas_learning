**VALUE COUNTS**

```python
import pandas as pd
```

```python
import kagglehub

path = kagglehub.dataset_download(
    "shivasingh4945/enhanced-breast-cancer-diagnostic-dataset"
)

print("Path to dataset files:", path)
```

```python
import os

print(os.listdir(path))
```

```python
df=pd.read_csv(f"{path}/breast_cancer_enhanced_dataset.csv")
```

1.BASIC IMPLEMENTATION

```python
df['diagnosis'].value_counts()
```

2.IN ASCENDING ORDER

```python
df['diagnosis'].value_counts(ascending=True)
```

3.RELATIVE FREQUENCY

```python
df['diagnosis'].value_counts(normalize=True)*100
```

4.DROP THE NULL ROWS

```python
df['diagnosis'].value_counts(dropna=True)
```

5.USING APPLY AND VALUE COUNTS TOGETHER

```python
df['area_mean'].apply(lambda x: "Dangerous" if x > 200 else "non dangerous").value_counts()
```
