```python
import pandas as pd
```

```python
import kagglehub

path = kagglehub.dataset_download(
    "shivasingh4945/enhanced-breast-cancer-diagnostic-dataset"
)

print("Path to dataset files:", path)
```

```python
import os

print(os.listdir(path))
```

```python
df=pd.read_csv(f"{path}/breast_cancer_enhanced_dataset.csv")
```

1.SINGLE COLUMN FILTER

```python
df[df['radius_mean']>=28]
```

2.MULTIPLE COLUMN FILTER

```python
df[(df['radius_mean'] >= 28) & (df['malignancy_risk_score'] >= 66)]
```

3.isin METHOD

```python
df[(df['diagnosis']).isin(['M'])]
```

```python
df[df['malignancy_risk_score'].isin([60, 67])]  #It checks whether there is any value that has 60 or 67
```

4.between METHOD

```python
df[df['malignancy_risk_score'].between(65, 67)] #It checks whether there are values in between the mentioned values
```

5.query METHOD

```python
df.query('radius_mean >= 28')
```

6.filter METHOD

```python
df.filter(regex='mean')
```

```python
df.filter(items=['id','malignancy_risk_score'])
```
