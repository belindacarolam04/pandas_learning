**DETECTING MISSING VALUES**

```python
import pandas as pd
```

```python
df=pd.read_csv(r"/content/Diabetes Missing Data.csv")
```

1.BASIC IMPLEMENTATION

```python
df.isnull()
```

2.NUMBER OF NULL VALUES

```python
df.isnull().sum()
```

3.TOTAL NUMBER OF NULL VALUES

```python
print(df.isnull().sum().sum())
```

4.ROWS WITH ATLEAST ONE NULL VALUES

```python
df[df.isnull().any(axis=1)].head(10)
```

5.ROWS WITH NULL VALUES IN ALL FIELDS

```python
df[df.isnull().all(axis=1)]
```

6.COLUMN NAMES WITH NULL VALUES

```python
df.columns[df.isnull().any()]
```

7.TO FIND WHETHER NULL VALUES PRESENT

```python
df.isnull().values.any()
```

8.PERCENTAGE OF NULL VALUES

```python
(df.isnull().sum()/len(df))*100
```

9.FILTERING MISSING ROWS IN ONE COLUMN

```python
df['Diabetes_Pedigree'].isnull().head(10)
```

10.FILTERING MISSING VALUES IN MULTIPLE COLUMNS

```python
df[['Diastolic_BP','Diabetes_Pedigree']].isnull().head(10)
```

11.NOT NULL IMPLEMENTATION

```python
df.notnull()
```

12.TOTAL SUM OF NOT NULL VALUES

```python
print(df.notnull().sum().sum())
```

13.TO CHECK WHETHER ALL VALUES ARE NULL VALUES

```python
df.notnull().all().all()
```
