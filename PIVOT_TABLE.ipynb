**PIVOT TABLE**

 similar to groupby but give excel pivot table like output

```python
import pandas as pd
```

```python
data = {
    "Department":["IT","HR","IT","Sales","HR","Sales"],
    "Employee":["A","B","C","D","E","F"],
    "Salary":[50000,40000,60000,45000,42000,48000],
    "Bonus":[5000,7000,3000,3500,2000,4000],
    "Gender":["Male","Female","Male","Female","Male","Female"]
}

df = pd.DataFrame(data)
print(df)
```

BASIC IMPLEMENTATION

```python
print(df.pivot_table(index="Department",values="Salary"))
```

AGGREGATION

```python
print(df.pivot_table(index="Department",values='Salary',aggfunc='sum'))
```

MULTIPLE AGGREGATION

```python
print(df.pivot_table(index="Department",values='Salary',aggfunc=['sum','mean','median','var']))
```

MULTIPLE VALUE COLUMNS

```python
print(df.pivot_table(index='Department',values=['Bonus','Salary']))
```

MULTIPLE INDEX  COLUMNS

```python
print(df.pivot_table(index=['Department','Gender'],values=['Salary']))
```

COLUMNS

```python
print(df.pivot_table(index='Department',values='Salary',columns='Gender'))
```

TOTAL AND TOTAL NAME

```python

print(df.pivot_table(index="Department",values='Salary',columns='Gender',margins=True,margins_name="Total"))
```

HANDLING MISSING VALUES

```python
print(df.pivot_table(index='Department',values='Salary',fill_value=0))
```
