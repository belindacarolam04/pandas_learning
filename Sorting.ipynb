**SORTING**

```python
import pandas as pd
```

```python
import kagglehub

path = kagglehub.dataset_download(
    "shivasingh4945/enhanced-breast-cancer-diagnostic-dataset"
)

print("Path to dataset files:", path)
```

```python
import os

print(os.listdir(path))
```

```python
df=pd.read_csv(f"{path}/breast_cancer_enhanced_dataset.csv")
```

1.SORT BASED ON ONE COLUMN

```python
df.sort_values('radius_mean').head(10)
```

2.SORT BASED ON MULTIPLE COLUMNS

```python
df.sort_values(['radius_mean','texture_mean']).head(10)
```

3.SORTING MULTIPLE COLUMNS IN DIFFERENT ORDER

```python
df.sort_values(['radius_mean','texture_mean'],ascending=[True,False]).head(10)
```

4.SORTING BASED ON INDEX

```python
df.sort_index().head(10)
```

5.DESCENDING ORDER

```python
df.sort_index(ascending=False).head(10) #descending
```

7.SORTING IN ALPHABETICAL ORDER

```python
df.sort_values('diagnosis').head(10) #arranges in alphabetical order
```

8.SORT BASED ON COLUMN NAMES

```python
df.sort_index(axis=1) #arranges the columns in alphabetical order
```

9.N LARGEST

```python
df.nlargest(8,'area_mean')
```

10.N SMALLEST

```python
df.nsmallest(8,'area_mean')[['id','area_mean']]
```
