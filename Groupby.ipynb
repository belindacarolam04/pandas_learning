**GROUPBY**

Functions:


*   Step:1 Split data into groups
*   Step:2 Apply Calculations


*   Step:3 Combine results





```python
import pandas as pd
```

```python
data={'Department':['IT','HR','IT','Sales','HR','Sales'],
      'Employee':['A','B','C','D','E','F'],
      'Gender':['F','M','M','F','F','M'],
      'Salary':[50000,40000,60000,45000,70000,55000]}
df=pd.DataFrame(data)
print(df)
```

AGGREGATION


*   mean
*   median


*   standard deviation

*   variance
*   maximum


*   minimum





```python
group=df.groupby("Department")
sal=group['Salary']
print("--------\nMean\n--------")
print(sal.mean())
print("--------\nMedian\n---------")
print(sal.median())
print("------------------\nStandard Deviation\n--------------------")
print(sal.std())
print("------------\nVariance\n-------------")
print(sal.var())
print("Maximum Salary:",sal.max())
print("Minimum Salary:",sal.min())
print("\nCount of members in each department:")
print(group['Employee'].count())
```

MULTIPLE AGGREGATION

```python
print(df.groupby('Department')['Salary'].agg(['sum','mean','max','min']))
```

With Custom Names

```python
print(df.groupby('Department').agg(
    Total_pay=('Salary','sum'),
    Average_Salary=('Salary','mean'),
    Maximum=('Salary','max'),
    Minimum=('Salary','min')
))
```

GROUPBY MULTIPLE COLUMNS

```python
print(df.groupby(['Department','Gender'])['Salary'].mean())
```

SELECT MULTIPLE COLUMNS

```python
print(df.groupby("Department")[
    ["Salary","Gender"]
].sum())
```

GET A SPECIFIC GROUP

```python
print(df.groupby('Gender').get_group('F'))
print(df.groupby('Department').get_group('Sales'))
```

GROUP NAMES AND NUMBER OF GROUPS

```python
print("Number of groups:",df.groupby('Department').ngroups)
print("Groups:",df.groupby('Department').groups)
```

ACCESSING ROWS OF THE GROUP

```python
gr=df.groupby('Department')['Salary']
print(gr.first())
print(gr.last())
print(gr.nth(1))
```

USING TRANSFORM FUNCTION

```python
print(gr.transform('mean'))
df["Dept_Avg"] = (
    df.groupby("Department")
      ["Salary"]
      .transform("mean")
)
print(df)
```

RANKING

```python
df['rank_order']=gr.rank(ascending=False)
print(df[['Department','rank_order']])
```

ITERATE THROUGH GROUPS

```python
for dept,group in df.groupby("Department"):
  print(dept)
  print(group)
```

MULTIPLE FUNCTIONS ON MULTIPLE COLUMNS

```python
print(df.groupby('Department').agg({"Salary":['sum','mean','max'],"Employee":'count'}))
```
