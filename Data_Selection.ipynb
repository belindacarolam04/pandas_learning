**DATA SELECTION**

```python
import pandas as pd
import numpy as np
```

```python
import kagglehub

path = kagglehub.dataset_download(
    "shivasingh4945/enhanced-breast-cancer-diagnostic-dataset"
)

print("Path to dataset files:", path)
```

```python
import os

print(os.listdir(path))
```

```python
df=pd.read_csv(f"{path}/breast_cancer_enhanced_dataset.csv")
```

**ROW SELECTION**

```python
df.dtypes
```

1.SELECT SINGLE ROW

```python
df.loc[888]
```

2.SELECT MULTIPLE ROWS

```python
df.loc[[889,900]]
```

3.ROW RANGE

```python
df.loc[800:810]
```

4.SELECT ONE VALUE

```python
print(df.loc[1000,'radius_mean'].round(2))
```

5.ENTIRE ROW

```python
df.loc[1,:]
```

6.UPDATE VALUES

```python
df.loc[1,'diagnosis']='M'
print(df.loc[1])
```

**COLUMN SELECTION**

1.ONE COLUMN

```python
df.loc[:,'area_mean']
```

2.MULTIPLE SELECTION

```python
df.loc[:,['id','diagnosis','malignancy_risk_score']]
```

**SPECIFIC ROW AND COLUMN SELECTION**

```python
df.loc[[4,9,6],['id','diagnosis','malignancy_risk_score']]
```

**CONDITIONAL SELECTION**

```python
df.loc[df['diagnosis']=='B']
```

**MULTIPLE CONDITION**

```python
df.loc[(df['diagnosis']=='M') & (df['radius_mean']>15)]
```
