**MERGING**

```python
import pandas as pd
```

```python
employees=pd.DataFrame({'EmpId':[1,2,3,4],
                        'Name':['Alice','Bob','Charlie','David']})
print(employees)
```

```python
salaries=pd.DataFrame({'EmpId':[1,2,3,5],
                       'Salary':[50000,60000,70000,80000]})
print(salaries)
```

TYPES OF MERGING

INNER JOIN-default

matching rows alone

```python
(employees.merge(salaries,on='EmpId',how='inner'))
```

LEFT JOIN

all rows from left and common between both

```python
(employees.merge(salaries,on='EmpId',how='left'))
```

RIGHT JOIN

rows from right and common in both

```python
employees.merge(salaries,on='EmpId',how='right')
```

OUTER JOIN

combines all rows from both dataframe

```python
employees.merge(salaries,on='EmpId',how='outer')
```

MERGING BASED ON DIFFERENT COLUMNS

```python
df1 = pd.DataFrame({
    "ID":[1,2,3]
})

df2 = pd.DataFrame({
    "EmpId":[1,2,3],
    "Salary":[50000,60000,70000]
})
```

```python
df1.merge(df2,left_on='ID',right_on='EmpId')
```

MERGE MULTIPLE COLUMNS

```python
df3=pd.DataFrame({'Department':['HR','IT','Sales'],
                  'Name':['Alice','Bob','Monisha'],
                  'City':['New York','San Francisco','Los Angeles']})
print(df3)
df4=pd.DataFrame({'Department':['HR','IT','Sales'],
                  'Salary':[50000,60000,70000],'City':['New York','San Franciso','Los Angeles']})
print(df4)
```

```python
df3.merge(df4,on=['Department','City'],how='outer')
```

SUFFIXES FOR DUPLICATE COLUMNS

```python
df3.merge(df4,on='Department',how='outer',suffixes=('_old','_new'))
```

MERGE BASED ON INDEX

```python
df3.merge(
    df4,
    left_index=True,
    right_index=True
)
```

MERGING THREE DATAFRAMES

```python
df5=pd.DataFrame({'EmpId':[1,2,3,4],

                  'Salary':[50000,60000,70000,40000]})
print(df5)
```

```python
df5.merge(employees,on="EmpId").merge(df2,on="EmpId")
```

CROSS JOIN

Every row combines with every rows

```python
df1.merge(df2,how='cross')
```

SOURCES OF ROWS

```python
employees.merge(salaries,on='EmpId',how='outer',indicator=True)
```

VALIDATING MERGE

```python
employees.merge(salaries,on='EmpId',how='outer',validate='many_to_one')
```
