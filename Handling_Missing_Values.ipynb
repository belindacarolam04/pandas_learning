```python
import pandas as pd
```

```python
df=pd.read_csv(r"/content/Diabetes Missing Data.csv")
```

**PERECENTAGE OF NULL VALUES**

```python
(df.isnull().sum()/len(df))*100
```

**DROPPING NULL VALUES**

---

when removing the data does not affect the whole dataset,we tend to drop the null values

1.BASIC IMPLEMENTATION

```python
df.dropna()
```

2.DROP IF ALL VALUES ARE NULL

```python
df.dropna(how='all')
```

3.DROP A COLUMN

Here we are dropping this column because it has null values nearly equal to 50%

```python
df.drop('Serum_Insulin',axis=1)
```

4.KEEP THE ROWS IF THE NULL VALUES ARE EQUAL TO OR GREATER THAN 'thresh'

```python
df.dropna(thresh=8)
```

5.SIMILARLY FOR COLUMNS

```python
df.dropna(axis=1,thresh=750)
```

6.DROP THE NULL VALUES OF SPECIFIC COLUMN

```python
df.dropna(subset=['Glucose'])
```

**DUPLICATE VALUES**

```python
df.duplicated().sum()
```

**FILLING MISSING VALUES**

1.FILL NULL WITH 0

```python
df.fillna(0)
```

2.FILL WITH STATISTICAL VALUES(mean,median,mode)

```python
df['Diastolic_BP'].fillna(df['Diastolic_BP'].mean())
```

3.FORWARD FILL

---
Use previous values


```python
df.ffill()
```

4.BACKWARD FILL

---
Use next value


```python
df.bfill()
```

5.FILL FIRST FIXED VALUES

```python
df['Skin_Fold'].fillna(df['Skin_Fold'].median(),limit=100)
```

6.FILL WITH DIFFERENT VALUES FOR DIFFERENT COLUMNS

```python
df.fillna({
    'Skin_Fold':df['Skin_Fold'].median(),
    'Diastolic_BP':df['Diastolic_BP'].mean()
})
```
